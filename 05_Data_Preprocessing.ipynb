# Data Preprocessing

## 1. Why Preprocess Data?

Raw data is often messy â€” it may contain:
- Missing values
- Categorical (non-numeric) features
- Values on very different scales

Preprocessing ensures our machine learning models train better and generalize well.

---

## 2. Key Preprocessing Techniques

### a. Standardization (Z-score)
Centers data around mean 0 with a standard deviation of 1.

### b. Normalization (Min-Max Scaling)
Scales all values between 0 and 1.

### c. Encoding Categorical Data
Convert text labels to numbers:
- `LabelEncoder`: for target labels
- `OneHotEncoder`: for features

---
