# Decision Trees

## 1. Introduction to Decision Trees

Decision Trees are a popular machine learning algorithm used for both classification and regression tasks. They work by splitting the data into branches based on feature values, making decisions in a tree-like structure.

---

## 2. Theory/Explanation

### What is a Decision Tree?
A decision tree uses a flowchart-like structure to make decisions. It asks a series of questions about the data, branching based on the answers, until it arrives at a prediction.

### Key Terms:
- **Root Node**: The first question or split in the tree.
- **Internal Nodes**: Decision points based on features.
- **Leaves**: Final predictions or outputs.
- **Entropy/Gini Impurity**: Metrics used to decide the best splits.

Decision Trees are easy to interpret but can overfit on training data, especially if not pruned.

---

## 3. Code Example

We'll train a Decision Tree on the Iris dataset.

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# Load the dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Train the Decision Tree
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

# Make predictions
predictions = clf.predict(X_test)

# Show predictions
print(predictions)


````python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Load and split data
iris = datasets.load_iris()
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Decision Tree
clf = DecisionTreeClassifier()
clf.fit(X, y)

# Plot tree
plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Decision Tree for Iris Dataset")
plt.show()
````
